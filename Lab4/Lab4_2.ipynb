```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler(delta, dt=0.001):

    a = 1
    b = 1
    
    tmax = 1.5/delta
    t = np.arange(0, tmax, dt)
    
    x = np.zeros(len(t))
    x[0] = delta
    
    for i in range(len(t)-1):
        
        dx = a*x[i]**2 - b*x[i]**3
        
        x[i+1] = x[i] + dt*dx
        
    return t, x
```

```python
deltas = [0.1 ,0.01 ,0.001 ]

plt.figure(figsize=(8,5))

for d in deltas:
    
    t, x = euler(d, dt=0.0001)
    
    plt.plot(t, x, label=f"δ = {d}")

plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("Solution of dx/dt = x² − x³")

plt.legend()
plt.grid()

plt.show()
```

# Numerical Integration of the Equation $\dot{x} = ax^2 - bx^3$

We consider the differential equation

$$
\dot{x} = ax^2 - bx^3, \qquad x(0) = \delta
$$

where $\delta$ is a small quantity. In this problem we take

$$
a = 1, \quad b = 1
$$

so the equation becomes

$$
\dot{x} = x^2 - x^3
$$

We numerically integrate the equation using Euler's method from

$$
t = 0 \quad \text{to} \quad t = \frac{1.5}{\delta}
$$

for three values of $\delta$:

$$
\delta = 0.1,\; 0.01,\; 0.001
$$

---

# Behaviour of the Solution

The numerical solution typically shows three distinct phases.

## Phase 1: Slow Growth

Initially,

$$
x(0) = \delta
$$

Since $\delta$ is very small,

$$
\dot{x} = x^2 - x^3 \approx x^2
$$

Thus

$$
\dot{x} \approx \delta^2
$$

which is extremely small. Therefore the solution increases very slowly at first.

---

## Phase 2: Sudden Ramp

At later times the solution suddenly increases very rapidly. This sharp increase occurs around

$$
t \sim \frac{1}{\delta}
$$

This behaviour appears in the numerical plots as a steep "ramp".

---

## Phase 3: Saturation

As $x$ approaches $1$, the equation becomes

$$
\dot{x} = x^2(1-x)
$$

When

$$
x \rightarrow 1
$$

we have

$$
\dot{x} \rightarrow 0
$$

so the solution approaches the stable equilibrium value

$$
x = 1
$$

---

# Why Does the Sudden Increase Occur?

To understand the sharp ramp, consider the equation when $x$ is small. In this region the cubic term is negligible:

$$
\dot{x} \approx x^2
$$

This equation can be solved analytically.

$$
\frac{dx}{dt} = x^2
$$

Separating variables:

$$
\frac{dx}{x^2} = dt
$$

Integrating,

$$
-\frac{1}{x} = t + C
$$

Using the initial condition

$$
x(0) = \delta
$$

we obtain

$$
x(t) = \frac{1}{\frac{1}{\delta} - t}
$$

This solution diverges when

$$
t = \frac{1}{\delta}
$$

Thus the solution grows extremely rapidly near

$$
t \sim \frac{1}{\delta}
$$

which explains the sharp increase observed in the numerical simulation.

---

# Why Must the Step Size Decrease for Smaller $\delta$?

When $\delta$ becomes smaller, the rapid growth occurs at a larger time scale:

$$
t \sim \frac{1}{\delta}
$$

Near this time the slope

$$
\dot{x}
$$

becomes very large.

Euler's method approximates the derivative using a finite step size $\Delta t$. If $\Delta t$ is too large, the method cannot accurately capture the steep change in the solution.

Therefore, for smaller values of $\delta$, the step size must be reduced in order to obtain a smooth and accurate numerical solution.

---

# Conclusion

The numerical solution shows a slow initial growth followed by a sharp increase around

$$
t \sim \frac{1}{\delta}
$$

and finally saturation near the equilibrium value $x = 1$.

This behaviour can be understood analytically by approximating the equation for small $x$, which reveals a blow-up time at $t = 1/\delta$. Smaller values of $\delta$ shift this rapid growth to later times and require smaller numerical step sizes to resolve the steep ramp accurately.
