```python
import numpy as np
import matplotlib.pyplot as plt

N = 30
x = np.linspace(0, N-1, 200)

a = 6
x0 = (N-1)/2  # symmetric center

# Compute c so endpoints are ~0
c = -a * np.cosh((0 - x0)/a)

# Catenary
y = a * np.cosh((x - x0)/a) + c

plt.plot(x, y, label="Analytical Catenary")
plt.scatter([0, N-1], [0, 0], color='red', label="Fixed Ends")
plt.xlabel("Mass index")
plt.ylabel("Height")
plt.title("Catenary with Fixed Endpoints")
plt.legend()
plt.show()
```



## Continuum Limit

In the limit $N \to \infty$, the chain becomes a smooth curve $y(x)$.

The energy functional becomes:
$$
E[y] = \int (T + \rho g y)\sqrt{1 + (y ')^2}\, dx
$$

---

## Euler–Lagrange Equation

Minimizing the functional gives:
$$
\frac{d}{dx}\left(\frac{\partial L}{\partial y'}\right) - \frac{\partial L}{\partial y} = 0
$$

Solving this leads to:

$$
y(x) = a \cosh\left(\frac{x - x_0}{a}\right) + c
$$

---

## Result

- The numerical solution obtained from energy minimization closely matches the analytical catenary curve.
- This validates the correctness of the discrete model.

---

## Conclusion

- The equilibrium shape of a hanging chain is a **catenary**.
- Numerical optimization provides an approximation to the analytical solution.
- As $N$ increases, the numerical solution converges to the analytical curve.
