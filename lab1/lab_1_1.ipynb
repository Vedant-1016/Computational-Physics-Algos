```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 1/(x+1)
```

Below given is the Taylor Series expansion of the function f(x)

```python
def taylor_series(x,a,n):
    T = np.zeros_like(x)
    for k in range(n):
        T += ((-1)**k / (a + 1)**(k + 1)) * (x - a)**k
    return T
    
        
```

```python
x = np.linspace(0, 1, 400)
a_values = [0.0, 0.5, 0.8]
n_values = [3, 5]
```

```python
for a in a_values:
    plt.figure(figsize=(6,4))
    plt.plot(x, f(x), 'k', linewidth=2, label='Exact f(x)')
    
    for n in n_values:
        plt.plot(x, taylor_series(x, a, n),
                 label=f'Taylor n={n}')
    
    plt.title(f"Taylor approximation about a = {a}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.show()
```
