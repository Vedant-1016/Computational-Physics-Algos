```python
import numpy as np
import matplotlib.pyplot as plt

def duffing(x, r, alpha, beta):
    return r - alpha*x - beta*x**3

x = np.linspace(-3, 3, 400)

# parameters
alpha = 1
beta = -1

for r in [-1, 0, 1]:
    dx = duffing(x, r, alpha, beta)
    plt.plot(x, dx, label=f"r={r}")

plt.axhline(0, color='black')
plt.legend()
plt.xlabel("x")
plt.ylabel("dx/dt")
plt.title("Duffing System Phase Plot")
plt.show()
```
