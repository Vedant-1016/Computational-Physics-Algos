```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Sine(x):
    return np.sin(x)
dt=1/100
t=np.arange(0,20,dt)

initial=[1,-1,2,-2]
plt.figure(figsize=(8,5))
for x0 in initial:
    x=np.zeros_like(t)
    x[0]=x0
    for i in range(1,len(x)):
        x[i]=x[i-1]+dt*Sine(x[i-1])
    plt.plot(t, x, label=f"x(0) = {x0}")

plt.axhline(0, linestyle="--", color="black", linewidth=1)
plt.xlabel("Time t")
plt.ylabel("x(t)")
plt.title("Euler trajectories for x' = sin(x)")
plt.legend()
plt.grid(True)
plt.show()

```
