```python
import numpy as np
import matplotlib.pyplot as plt

# Time setup
dt = 0.01
t = np.arange(0, 20*np.pi, dt)
```

```python
def solve_oscillator(omega):
    x = 0
    v = 0

    xs = []

    for ti in t:
        a = np.cos(omega * ti) - x

        x += v * dt
        v += a * dt

        xs.append(x)

    return np.array(xs)
```

```python
omegas = [0.5, 1.0, 2.0]

plt.figure()

for omega in omegas:
    x = solve_oscillator(omega)
    plt.plot(t, x, label=f"ω = {omega}")

plt.title("Response for Different Driving Frequencies")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.legend()
plt.grid()
plt.show()
```

# Forced Oscillator without Damping

## Equation

$$
\ddot{x} + x = \cos(\omega t)
$$

with initial conditions:
$$
x(0) = 0, \quad \dot{x}(0) = 0
$$

---

## General Solution

The solution consists of two parts:

$$
x(t) = x_{\text{hom}} + x_{\text{part}}
$$

---

### Homogeneous Solution

$$
x_{\text{hom}} = A\cos t + B\sin t
$$

---

### Particular Solution

For $\omega \neq 1$:

$$
x_{\text{part}} = \frac{\cos(\omega t)}{1 - \omega^2}
$$

---

## Complete Solution

$$
x(t) = A\cos t + B\sin t + \frac{\cos(\omega t)}{1 - \omega^2}
$$

---

## Resonance Case ($\omega = 1$)

When $\omega = 1$, the denominator vanishes, and the solution becomes:

$$
x(t) = \frac{t}{2} \sin t
$$

---

## Interpretation

### Case 1: $\omega < 1$
- System oscillates with bounded amplitude
- Driven frequency dominates slowly

---

### Case 2: $\omega \approx 1$ (Resonance)
- Amplitude grows with time
- Energy continuously increases
- No damping → unbounded growth

---

### Case 3: $\omega > 1$
- Oscillations remain bounded
- Faster oscillations with smaller amplitude

---

## Conclusion

- The system shows resonance at $\omega = 1$
- Unlike damped systems, amplitude grows indefinitely at resonance
- Numerical results match analytical predictions
